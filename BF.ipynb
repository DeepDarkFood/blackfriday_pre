```python
age = data.groupby(['User_ID','Age'])
c_age = age.count().count(level='Age')
```

```python
gender_sum_user = data.groupby(['User_ID','Gender']).sum()
gender_sum_user = gender_sum_user.sum(level='Gender')
plt.figure()
plt.pie(gender_sum_user['Purchase'],labels=['F','M'])
```
